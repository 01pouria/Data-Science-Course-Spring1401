```python
def f(x):
  return x
def g(x):
  return x**2
def h(x):
  return x**(4/3)
```

```python
x=[]
for i in range(10):
  x.append(i)
y=[]
for i in x:
  y.append(f(i))
y1=[]
for i in x:
  y1.append(g(i))
y2=[]
for i in x:
  y2.append(h(i))
```

```python
print(y)
print(y1)
print(y2)
```

```python
plt.plot(x,y,label='f(x)=x')
plt.scatter(x,y)
plt.plot(x,y1,label='g(x)=x^2')
plt.scatter(x,y1)
plt.plot(x,y2,label='h(x)=x^4/3')
plt.scatter(x,y2)
plt.title('functions plot')
plt.xlabel('input')
plt.ylabel('output')
plt.legend()
plt.grid()
plt.show()
```
