```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=pd.read_csv('/content/drive/MyDrive/x.csv')
y=pd.read_csv('/content/drive/MyDrive/y.csv')
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split as tts 
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression 
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
```

```python
print('train score : ',reg.score(x_train,y_train))
print('test score : ',reg.score(x_test,y_test))
```

```python
plt.scatter(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error as MSE 
mse=MSE(y_pred,y_test)
mse
```

```python
n=np.arange(1,127,1)
train_ac=np.empty(len(n))
test_ac=np.empty(len(n))
mse=np.empty(len(n))
```

```python
for i,k in enumerate(n):
  reg=LinearRegression()
  pca=PCA(n_components=k)
  pcamodel=pca.fit_transform(x)
  x_train,x_test,y_train,y_test=tts(pcamodel,y,test_size=0.2,random_state=42)
  reg.fit(x_train,y_train)
  y_pred=reg.predict(x_test)
  train_ac[i]=reg.score(x_train,y_train)
  test_ac[i]=reg.score(x_test,y_test)
  mse[i]=MSE(y_test,y_pred)
```

```python
plt.figure(figsize=(10,5),dpi=100)
plt.plot(n,train_ac,label='Train')
plt.plot(n,test_ac,label='Test')
plt.xlabel('components')
plt.ylabel('accuracy')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5),dpi=100)
plt.plot(n,mse,label='Train')
plt.bar(n,mse,label='Test')
plt.xlabel('components')
plt.ylabel('mse')
plt.legend()
plt.grid()
plt.show()
```

```python
df=pd.read_csv('/content/drive/MyDrive/ecg.csv')
df
```

```python
ll=[]
for i in range(140):
  ll.append('gate'+str(i))
ll.append('target')
```

```python
df.columns=ll
```

```python
df
```

```python
df1=df.drop('target',axis=1)
```

```python
from sklearn.cluster import DBSCAN
c=DBSCAN(eps=2,min_samples=3).fit(df1)
```

```python
l=c.labels_
```

```python
m=np.unique(l)
count=0
for i in m:
  count+=1
count
```

```python
def count(label):
  m=np.unique(label)
  count=0
  for i in m:
    count+=1
  return count
```

```python
count(l)
```

```python
eps=np.linspace(0.1,10,300)
n_s=np.linspace(1,100,300)
cluster=[]
for i in eps:
  for j in n_s:
    cl=DBSCAN(eps=i,min_samples=j).fit(df1)
    l=cl.labels_
    c=count(l)
    cluster.append([i,j,c])
```
