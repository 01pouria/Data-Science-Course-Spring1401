```python
e1=di['E1'].values
np.var(e1),np.std(e1),np.mean(e1)
```

```python
from sklearn.preprocessing import scale,normalize,minmax_scale
e1_n=scale(e1)
np.var(e1_n),np.std(e1_n),np.mean(e1_n)
```

```python
e1=e1.reshape(-1,1)
e1_n1=normalize(e1,norm='l1')
np.var(e1_n1),np.std(e1_n1),np.mean(e1_n1)
```

```python
e1=e1.reshape(-1,1)
e1_n1=normalize(e1,norm='l2')
np.var(e1_n1),np.std(e1_n1),np.mean(e1_n1)
```

```python
e1_n2=minmax_scale(e1,feature_range=(0,1))
np.var(e1_n2),np.std(e1_n2),np.mean(e1_n2)
```

```python
e1_n2=minmax_scale(e1,feature_range=(2,6))
np.var(e1_n2),np.std(e1_n2),np.mean(e1_n2)
```
